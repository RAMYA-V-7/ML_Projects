# EDA(Exploratory Data Analysis)
➡️It is used to explore,clean data and summarize the key insights.

```python
#importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#1.Loading data to jupyter notebook
df = pd.read_csv('movies.csv')
```

```python
#2.Length of dataset before cleaning
len(df)
```

```python
#3.Removing empty cells - removing rows which contains empty cells
df=df.dropna()
```

```python
df
```

```python
#4.Top 5 datas of dataset before cleaning
df.head()
```

```python
#5.Converting data to correct format
df['GENRE'] = df['GENRE'].str.replace(r'\n',"")
df['ONE-LINE'] = df['ONE-LINE'].str.replace(r'\n',"")
df['STARS'] = df['STARS'].str.replace(r'\n',"")
df['Gross'] = df['Gross'].str.replace(r'[$M]',"")
df['YEAR'] = df['YEAR'].str.replace(r'[()I-]',"")
```

```python
df['YEAR'] = df['YEAR'].str.replace(r'[â€“]',"")
```

```python
df['STARS'] = df['STARS'].str.replace(r'[|:?]',"")
df['ONE-LINE'] = df['ONE-LINE'].str.replace(r'[|"",?]',"")
```

```python
df
```

```python
#6.Discovering duplicates
df.duplicated()
```

```python
df['YEAR'] = pd.to_numeric(df['YEAR'])
```

```python
#7.Sum of null values
df.isnull().sum()
```

```python
df['Gross'] = pd.to_numeric(df['Gross'])
```

```python
#8.Resetting index for the cleaned dataset
df = df.reset_index(drop=True)
```

```python
#9.Length of cleaned dataset
len(df)
```

```python
#10.Applying styles to the dataset
df.style.set_properties(**{'background-color': 'black',
                           'color': 'purple'})
```

```python
#10.Top 5 datas of the cleaned dataset
df.head()
```

```python
#11.Applying groupby function on the dataset
df.groupby('MOVIES').agg({'RATING':'max'})
```

```python
#12.Sorting values with respect to columns for the dataset
df.sort_values(['VOTES'],ascending=False).reset_index(drop=True)
```

```python
#13.Renaming appropriate name for the columns of the cleaned dataset
df.rename(columns = {'MOVIES':'Movies','YEAR':'Year','GENRE':'Genre','RATING':'Rating','ONE-LINE':'One-Line Description','STARS':'Stars','RunTime':'Run Time','VOTES':'Votes'},inplace=True)
```
